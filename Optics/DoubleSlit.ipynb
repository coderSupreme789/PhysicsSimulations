# Double Slit

## Load libraries

```python
# Load libraries
import numpy as np

# Graphics / plotting
import matplotlib.pyplot as plt
```

## Parameters

```python
# Distance between the aperture and the screen
screen_distance = 100

# Wavelength of the incident light
wavelength = 0.1

# Distance between the slits
slit_separation = 0.1

# Intensity of light incident on the slits
slit_intensity = 100
```

## Intensity profiles

```python
# Intensity of a particle at some distance r
def intensity_profile(distance, kind="spherical"):
    if kind == "spherical":
        return 1 / (4 * np.pi * distance**2)
    if kind == "cylindrical":
        return 1 / (2 * np.pi * distance)
    if kind == "flat":
        return 1
```

## Intensity on the screen

```python
# Calculate intensity at given displacement on the screen
def intensity(deviation, kind="spherical"):

    # Calculate the left and right distances
    left_distance = np.sqrt(
        (screen_distance) ** 2 + (slit_separation / 2 - deviation) ** 2
    )
    right_distance = np.sqrt(
        (screen_distance) ** 2 + (slit_separation / 2 + deviation) ** 2
    )

    # Calculate amplitudes - this is wrong I'm pretty sure
    left_amplitude = intensity_profile(left_distance, kind) * np.cos(left_distance * 2 * np.pi / wavelength)
    right_amplitude = intensity_profile(right_distance, kind) * np.cos(right_distance * 2 * np.pi / wavelength)

    # Return the intensity at the screen at given position
    return (left_amplitude + right_amplitude) ** 2
```

## Plot the resulting intensity pattern

```python
# Plotting
deviations = np.linspace(-10, 10, 10000)
plt.plot(deviations, intensity(deviations, "flat"))
plt.show()
```
